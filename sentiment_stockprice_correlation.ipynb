##Semantic Analysis

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.downloader.download('vader_lexicon')
```

```python
new_words = {
    'red': -1.0,
    'bull': 1.0,
    'bear': -1.0
}
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer.lexicon.update(new_words)
```

####Calc Semantic scores of news title and content

```python
lis = [analyzer.polarity_scores('Bull market'),
analyzer.polarity_scores('Bear market')]
lis
```

```python
scores_title = df1['title'].apply(analyzer.polarity_scores).tolist()
scores_content = df1['content'].apply(analyzer.polarity_scores).tolist()
scores_title
```

```python
temp = pd.merge(df1['title'] , pd.DataFrame(scores_title) , left_index =True,right_index= True)
temp.tail(5).reset_index(drop=True)
```

```python
datelist = df1['release_date'].reset_index()
```

```python
df_scores_titles = pd.DataFrame(scores_title)
df_score_content = pd.DataFrame(scores_content)
```

```python
df_score_content.rename(columns = {'compound':'content_compound'}, inplace = True)
```

```python
df_scores_titles.rename(columns = {'compound':'titles_compound'}, inplace = True)
```

```python
compound_sm = pd.merge(df_scores_titles['titles_compound'] , df_score_content['content_compound'] , left_index =True,right_index= True)
compound_sm_np = pd.merge(compound_sm , df1['category'] , left_index =True,right_index= True)
```

```python
semantic_news = pd.merge(compound_sm_np , datelist['release_date'] , left_index =True,right_index= True)
semantic_news.sample(n=5)
```

```python
semantic_news['compound0_1'] = 0*semantic_news['titles_compound'] + 1*semantic_news['content_compound']
semantic_news['compound.1_.9'] = 0.1*semantic_news['titles_compound'] + 0.9*semantic_news['content_compound']
semantic_news['compound.2_.8'] = 0.2*semantic_news['titles_compound'] + 0.8*semantic_news['content_compound']
semantic_news['compound.3_.7'] = 0.3*semantic_news['titles_compound'] + 0.7*semantic_news['content_compound']
semantic_news
```

```python
semantic_news2 = semantic_news
semantic_news2.drop('titles_compound', axis=1, inplace=True)
semantic_news2.drop('content_compound', axis=1, inplace=True)
semantic_news_opinion = semantic_news2[semantic_news2['category'] == 'opinion']
semantic_news_news = semantic_news2[semantic_news2['category'] == 'news']
```

news series

```python
semantic_news_news2 = semantic_news_news.groupby(by=['release_date']).sum()
semantic_news_news3 = semantic_news_news.groupby(by=['release_date']).count()
semantic_news_news3 = semantic_news_news3.rename(columns = {
    'compound.3_.7' : 'count'
})
semantic_news_newsf = pd.merge(semantic_news_news2,semantic_news_news3['count'],right_index= True,left_index= True)
semantic_news_newsf = semantic_news_newsf.div(semantic_news_newsf['count'], axis=0)
semantic_news_newsf
```

opinion series

```python
semantic_news_opinion2 = semantic_news_opinion.groupby(by=['release_date']).sum()
semantic_news_opinion3 = semantic_news_opinion.groupby(by=['release_date']).count()
semantic_news_opinion3 = semantic_news_opinion3.rename(columns = {
    'compound.3_.7' : 'count'
})
semantic_news_opinionf = pd.merge(semantic_news_opinion2,semantic_news_opinion3['count'],right_index= True,left_index= True)

semantic_news_opinionf = semantic_news_opinionf.div(semantic_news_opinionf['count'], axis=0)
semantic_news_opinionf
```

combining news and opinion

```python
cnp = pd.merge(semantic_news_newsf,semantic_news_opinionf,on='release_date',how='outer')
cnp = cnp.fillna(0)
cnp
```

```python
cnp['compound0_1_x']
```

```python
mulx = 0.7
muly = 0.3
```

```python
cnp['compound0_1'] = mulx*cnp['compound0_1_x'] + muly*cnp['compound0_1_y']
cnp['compound.1_.9'] = mulx*cnp['compound.1_.9_x'] + muly*cnp['compound.1_.9_y']
cnp['compound.2_.8'] = mulx*cnp['compound.2_.8_x'] + muly*cnp['compound.2_.8_y']
cnp['compound.3_.7'] = mulx*cnp['compound.3_.7_x'] + muly*cnp['compound.3_.7_y']
```

```python
cnpf = cnp[['compound0_1', 'compound.1_.9', 'compound.2_.8', 'compound.3_.7']]
cnpf
```

```python
cnpf['compound.1_.9'] = cnpf['compound.1_.9'].abs()

## Get Max compund
# final_df = semantic_fac.loc[semantic_fac['compound'].idxmax()]
# news_date = semantic_fac[semantic_fac['compound'] == final_df['compound']]

## Get Max 3
cnpf = cnpf.sort_values('compound.1_.9')
final_df_h = cnpf.tail(3)
final_df_l = cnpf.head(3)

final_df_h = final_df_h.sort_values('release_date')
final_df_l = final_df_l.sort_values('release_date')
final_df_l
```

Lag plot :
Lag plots are used to check if a data set or time series is random. Random data should not exhibit any structure in the lag plot. Non-random structure implies that the underlying data are not random. 

```python
from pandas.plotting import lag_plot

plt.figure();

spacing = np.linspace(-99 * np.pi, 99 * np.pi, num=1000)

lag_plot(cc['compound.3_.7']);
```

autocorrelation_plot : checking randomness in time series.
if a series is random, such autocorrelations should be near zero for any and all time-lag separations. If time series is non-random then one or more of the autocorrelations will be significantly non-zero. 

```python
from pandas.plotting import autocorrelation_plot

plt.figure();

spacing = np.linspace(-9 * np.pi, 9 * np.pi, num=1000)

autocorrelation_plot(cc['compound.3_.7']);
```

Bootstrap plot:
Bootstrap plots are used to visually assess the uncertainty of a statistic, such as mean, median, midrange, etc. A random subset of a specified size is selected from a data set, the statistic in question is computed for this subset and the process is repeated a specified number of times. Resulting plots and histograms are what constitutes the bootstrap plot.

```python
from pandas.plotting import bootstrap_plot

bootstrap_plot(cc['compound.3_.7'], size=50, samples=500, color="grey");
```
